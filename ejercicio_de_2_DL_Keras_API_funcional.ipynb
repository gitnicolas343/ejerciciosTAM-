# Creación de arquitecturas de aprendizaje profundo mediante API funcional de Keras.

- Se presentan los consideraciones básicas respecto a la [API funcional de Keras](https://keras.io/guides/functional_api/) para la implementación de modelos de redes profundas con TensorFlow.



```
import tensorflow as tf #importar tensorflow
print(tf.__version__)
```

- Se utilizará la base de datos Fashion Mnist nuevamente, normalizando las imágenes de 0 a 1.

```
train, test = tf.keras.datasets.fashion_mnist.load_data()
images, labels = train
```

- Se presenta un ejemplo de la base de datos

```
import matplotlib.pyplot as plt
plt.imshow(images[5000,:,:],cmap= 'binary')
plt.show()
```

```
import numpy as np
print(np.unique(labels)) #etiquetas
print(images.shape) #tamaño de las imágenes
```

```
Xtrain, ytrain = train
Xtrain = Xtrain/255 # tipo flotante, normalizados de 0 a 1
Xtest, ytest = test
Xtest = Xtest/255 # tipo flotante, normalizados de 0 a 1
```

- En el cuaderno introductorio a redes neuronales con tensorflow se planteó la construcción de modelos mediante secuencia de capas:

```
#definir arquitectura secuencial
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(Xtrain.shape[1],Xtrain.shape[2])),
    tf.keras.layers.Dense(200,activation="relu", name='red1'),
    tf.keras.layers.Dense(50, activation="tanh",name='hred2'),
    tf.keras.layers.Dense(10,activation='softmax',name='output')
])

model.summary() # resumen del modelo
tf.keras.utils.plot_model(model) #diagrama del modelo
```

- Aunque la implementacióm mediante secuencia es intuitiva, no permite realizar conexiones flexibles entre capas.

- En ese sentido, Keras permite crear capas bajo el principio de funciones y argumentos, como se muestra a continuación.

**Nota**: a diferencia de la estructura secuencial, en la API funcional se debe indicar la capa o capas de entrada y la capa o capas de salida para crear el grafo computacional de forma apropiada.

```
#arquitectura funcional
#capa de entrada
input_l = tf.keras.layers.Input(shape=(Xtrain.shape[1],Xtrain.shape[2]), name='entrada')
#capa de aplanamiento de las imágenes
flatten = tf.keras.layers.Flatten(input_shape=(Xtrain.shape[1],Xtrain.shape[2]))(input_l)#argumento de entrada
#capas densas
h1 = tf.keras.layers.Dense(200,activation='tanh',name='h1')(flatten)#argumento de entrada
h2 = tf.keras.layers.Dense(50,activation='tanh',name='h2')(h1)
#capa de salida
output = tf.keras.layers.Dense(10,activation="softmax",name='output')(h2)
#crear modelo según conexiones tipo funcionales
model_fun = tf.keras.Model(inputs=input_l,outputs=output)

#resumen y gráfica del modelo
model_fun.summary()
tf.keras.utils.plot_model(model_fun)
```

# La API funcional permite realizar conexiones más flexibles

- En el siguiente ejemplo se crea un modelo con una entrada y dos salidas con interconexiones flexibles:

```
#arquitectura funcional 2
tf.keras.backend.clear_session()
input_l = tf.keras.layers.Input(shape=(Xtrain.shape[1],Xtrain.shape[2]), name='entrada')
flatten = tf.keras.layers.Flatten(input_shape=(Xtrain.shape[1],Xtrain.shape[2]))(input_l)

h1 = tf.keras.layers.Dense(200,activation='tanh',name='h1')(flatten)
h2 = tf.keras.layers.Dense(50,activation='tanh',name='h2')(h1)
#capa que concatena caracterísitacas extraídas en h2 con imágenes de entrada
concat = tf.keras.layers.concatenate([h2,flatten])
#se crean dos salidas:
output_A = tf.keras.layers.Dense(10,activation="softmax",name='outputA')(concat)#desde concatenación
output_B = tf.keras.layers.Dense(10,activation="softmax",name='outputB')(h2)#desde h2
#las entradas y salidas se ingresan como listas
model_fun = tf.keras.Model(inputs=input_l,outputs=[output_A, output_B])
#model_fun.summary()
tf.keras.utils.plot_model(model_fun)
```

**Nota**: si se cuenta con varias salidas se puede definir un mismo costo para cada salida, o una lista con los costos requeridos. Además, en la lista `loss_weights` se puede especificar el valor del peso asociado al costo de cada salida en el intervalo [0,1].

```
model_fun.compile(loss="sparse_categorical_crossentropy", #custom_loss(),#custom_loss(),#"sparse_categorical_crossentropy",
              optimizer="Adam",
              metrics=["accuracy","accuracy"]) #f1, precision, recall, crossentropy
```

- El entrenamiento se realiza de igual forma al utilizado con el modelo secuencial.

- Como se fijaron dos salidas, se debe ingresar una lista para los argumentos ytrain.

```
history = model_fun.fit(Xtrain, [ytrain,ytrain] , epochs=10,batch_size=64, # 32, 64, 128, 256
                    validation_split=0.3) # se fija el conjunto de validación como el 20% del de train
```

- Se presentan las curvas de desempeño:

```
import pandas as pd

pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1)
#save_fig("keras_learning_curves_plot")
plt.show()
```

```
#plot mnist 2D
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
def plot_mnist_2d(Z,y,images,img_w=28,img_h=28,zoom=0.5,cmap='jet'):
    fig, ax = plt.subplots(figsize=(16,10))
    for i in range(Z.shape[0]):
        #print('img',i+1,'/',Z.shape[0])
        image = images[i].reshape((img_w, img_h))
        im = OffsetImage(image, zoom=zoom,cmap=cmap)
        ab = AnnotationBbox(im, (Z[i,0], Z[i,1]), xycoords='data', frameon=False)
        ax.add_artist(ab)
        ax.update_datalim([(Z[i,0], Z[i,1])])
        ax.autoscale()
```

```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Cargar datos MNIST
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

# Preprocesamiento
X_train = X_train[..., np.newaxis] / 255.0
X_test = X_test[..., np.newaxis] / 255.0

# Generar versiones con ruido
def add_noise(images, var):
    noise = np.random.normal(0, var**0.5, images.shape)
    noisy = np.clip(images + noise, 0, 1)
    return noisy

X_train_noise1 = add_noise(X_train, 0.01)
X_train_noise2 = add_noise(X_train, 0.25)
X_test_noise1 = add_noise(X_test, 0.01)
X_test_noise2 = add_noise(X_test, 0.25)

# Etiquetas de paridad
y_train_parity = y_train % 2
y_test_parity = y_test % 2
```

```
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Concatenate
from tensorflow.keras.models import Model

# Función para crear una rama convolucional
def create_cnn_branch(input_layer):
    x = Conv2D(32, (3,3), activation='relu')(input_layer)
    x = MaxPooling2D((2,2))(x)
    x = Conv2D(64, (3,3), activation='relu')(x)
    x = MaxPooling2D((2,2))(x)
    x = Flatten()(x)
    return x

# Definir entradas
input_orig = Input(shape=(28,28,1), name='input_original')
input_noise1 = Input(shape=(28,28,1), name='input_noise1')
input_noise2 = Input(shape=(28,28,1), name='input_noise2')

# Procesar cada entrada
branch_orig = create_cnn_branch(input_orig)
branch_noise1 = create_cnn_branch(input_noise1)
branch_noise2 = create_cnn_branch(input_noise2)

# Concatenar características
concat = Concatenate()([branch_orig, branch_noise1, branch_noise2])

# Capas densas
dense = Dense(128, activation='relu')(concat)

# Salidas
output_digit = Dense(10, activation='softmax', name='output_digit')(dense)
output_parity = Dense(1, activation='sigmoid', name='output_parity')(dense)

# Modelo final
model = Model(inputs=[input_orig, input_noise1, input_noise2],
              outputs=[output_digit, output_parity])

model.summary()
tf.keras.utils.plot_model(model, show_shapes=True)
```

```
model.compile(optimizer='adam',
              loss={'output_digit': 'sparse_categorical_crossentropy',
                    'output_parity': 'binary_crossentropy'},
              metrics={'output_digit': 'accuracy',
                       'output_parity': 'accuracy'})

history = model.fit(
    [X_train, X_train_noise1, X_train_noise2],
    [y_train, y_train_parity],
    epochs=10,
    batch_size=64,
    validation_data=(
        [X_test, X_test_noise1, X_test_noise2],
        [y_test, y_test_parity]
    )
)
```

```
# Evaluación
test_results = model.evaluate(
    [X_test, X_test_noise1, X_test_noise2],
    [y_test, y_test_parity]
)
print(f"Test loss (digit): {test_results[1]}, Test accuracy (digit): {test_results[3]}")
print(f"Test loss (parity): {test_results[2]}, Test accuracy (parity): {test_results[4]}")

# Curvas de aprendizaje
plt.figure(figsize=(12, 4))
plt.subplot(1,2,1)
plt.plot(history.history['output_digit_accuracy'], label='Train Digit')
plt.plot(history.history['val_output_digit_accuracy'], label='Validation Digit')
plt.title('Digit Classification Accuracy')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['output_parity_accuracy'], label='Train Parity')
plt.plot(history.history['val_output_parity_accuracy'], label='Validation Parity')
plt.title('Parity Classification Accuracy')
plt.legend()
plt.show()
```

```
# Modelo intermedio para extraer características
feature_extractor = Model(inputs=model.inputs, outputs=concat)

# Extraer características del conjunto de prueba
features = feature_extractor.predict([X_test, X_test_noise1, X_test_noise2])

# PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
features_pca = pca.fit_transform(features)

plt.scatter(features_pca[:,0], features_pca[:,1], c=y_test)
plt.colorbar()
plt.title('PCA de Características Combinadas')
plt.show()

# t-SNE (submuestra para eficiencia)
from sklearn.manifold import TSNE
subset = np.random.choice(len(features), 1000, replace=False)
tsne = TSNE(n_components=2)
features_tsne = tsne.fit_transform(features[subset])

plt.scatter(features_tsne[:,0], features_tsne[:,1], c=y_test[subset])
plt.colorbar()
plt.title('t-SNE de Características Combinadas')
plt.show()
```

```
# Evaluación del modelo en el conjunto de prueba
test_results = model.evaluate(
    [X_test, X_test_noise1, X_test_noise2],
    [y_test, y_test_parity],
    verbose=0
)

# Imprimir resultados detallados
print("\nResultados de evaluación en conjunto de prueba:")
print(f"Pérdida total: {test_results[0]:.4f}")
print(f"Pérdida clasificación dígitos: {test_results[1]:.4f}")
print(f"Pérdida clasificación paridad: {test_results[2]:.4f}")
print(f"Exactitud dígitos: {test_results[3]:.4f}")
print(f"Exactitud paridad: {test_results[4]:.4f}")

# Generar predicciones
y_pred_digit, y_pred_parity = model.predict([X_test, X_test_noise1, X_test_noise2])

# Métricas para clasificación de dígitos
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Convertir predicciones a clases
y_pred_digit_classes = np.argmax(y_pred_digit, axis=1)

print("\nReporte de clasificación para dígitos:")
print(classification_report(y_test, y_pred_digit_classes))

plt.figure(figsize=(10,8))
sns.heatmap(confusion_matrix(y_test, y_pred_digit_classes),
            annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de confusión para clasificación de dígitos')
plt.xlabel('Predicho')
plt.ylabel('Real')
plt.show()

# Métricas para clasificación de paridad
y_pred_parity_binary = (y_pred_parity > 0.5).astype(int)

print("\nReporte de clasificación para paridad:")
print(classification_report(y_test_parity, y_pred_parity_binary))

plt.figure(figsize=(6,6))
sns.heatmap(confusion_matrix(y_test_parity, y_pred_parity_binary),
            annot=True, fmt='d', cmap='Greens')
plt.title('Matriz de confusión para clasificación de paridad')
plt.xlabel('Predicho')
plt.ylabel('Real')
plt.show()

# Visualizar algunas predicciones
sample_indices = np.random.choice(len(X_test), 5, replace=False)

plt.figure(figsize=(15, 5))
for i, idx in enumerate(sample_indices, 1):
    plt.subplot(1, 5, i)
    plt.imshow(X_test[idx].squeeze(), cmap='gray')
    plt.title(f"Real: {y_test[idx]}\nPred: {y_pred_digit_classes[idx]}\nParidad: {'Par' if y_pred_parity_binary[idx][0] else 'Impar'}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```
